```python
X_train=pd.read_csv('K:\Coding Blocks\Machine learning\Logistic_X_Train.csv')
Y_train=pd.read_csv('K:\Coding Blocks\Machine learning\Logistic_Y_Train.csv')

X_test=pd.read_csv('K:\Coding Blocks\Machine learning\Logistic_X_Test.csv')
```

```python

X_train=X_train.values
Y_train=Y_train.values
X_test=X_test.values
```

```python
u=X_train.mean()
std=X_train.std()
#print(u,std)
X_train=(X_train-u)/std
```

```python
def hypothesis(x,w,b):
    '''accepts input vector x, input weight vector w and bias b'''
    
    h = np.dot(x,w) + b
    return sigmoid(h)

def sigmoid(z):
    return 1.0/(1.0 + np.exp(-1.0*z))

def error(y_true,x,w,b):
    
    m = x.shape[0]
    
    err = 0.0
    
    for i in range(m):
       
        hx = hypothesis(x[i],w,b) 
        err += y_true[i]*np.log2(hx) + (1-y_true[i])*np.log2(1-hx)
        
    
    return -err/m


def get_grads(y_true,x,w,b):
    
    grad_w = np.zeros(w.shape)
    grad_b = 0.0
    
    m = x.shape[0]
    
    for i in range(m):
        hx = hypothesis(x[i],w,b)
        
        grad_w += (y_true[i] - hx)*x[i]
        grad_b +=  (y_true[i]-hx)
        
    
    grad_w /= m
    grad_b /= m
    
    return [grad_w,grad_b]


# One Iteration of Gradient Descent
def grad_descent(x,y_true,w,b,learning_rate=0.1):
    
    err = error(y_true,x,w,b)
    [grad_w,grad_b] = get_grads(y_true,x,w,b)
    
    w = w + learning_rate*grad_w
    b = b + learning_rate*grad_b
    
    return err,w,b
    
def predict(x,w,b):
    
    confidence = hypothesis(x,w,b)
    if confidence<0.5:
        return 0
    else:
        return 1
    
def y_predict(x_tst,w,b):
    
    y_pred = []
    
    for i in range(x_tst.shape[0]):
        p = predict(x_tst[i],w,b)
        
        y_pred.append(p)
        
    y_pred = np.array(y_pred)
    
    
    return y_pred 
```

```python
acc = []
loss = []

W = 2*np.random.random((X_train.shape[1],))

b = 5*np.random.random()
```

```python

for i in range(1000):
    l,W,b = grad_descent(X_train,Y_train,W,b,learning_rate=0.1)
    acc.append(y_predict(X_test,W,b))
    loss.append(l)
```

```python
plt.plot(loss)
plt.ylabel("Negative of Log Likelihood")
plt.xlabel("Time")
plt.show()
```
